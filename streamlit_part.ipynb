## Application web



**In this part of our project, we combined all of our Streamlit functionality into a single `app.py` file.** By doing so, we only need to run Streamlit once to access our entire user interface, allowing us to seamlessly visualize and interact with our analysis results.

The application features **three main pages**:

1. **Global Dashboard:** Provides an overview of tourism trends over time, including key metrics such as the latest arrivals and overall growth.
A detailed chart also allows users to select a specific country (e.g., Argentina, Brazil, or Colombia) to observe spikes, dips, or other notable changes in tourism‚Äîand potentially link those trends to any reported negative impacts.

2. **Sentiment Analysis:** Lets users analyze any chosen article to determine whether its sentiment is positive, negative, or neutral.

3. **Summarization:** Summarizes lengthy articles using a T5-based model, helping users quickly grasp the main ideas without reading through the entire text.

If we go in detailed on the code, we implemented for example, a **`TextCleaner` class** that removes stop words, URLs, email addresses, HTML tags, punctuation, and numbers. This ensures a standardized, noise-free dataset before further analysis.

Next, the **`TourismChatbot` class** uses a T5-based model from Hugging Face to summarize articles. This summarization is especially helpful for sentiment classification, as condensing lengthy articles often yields more accurate results.

Finally, the application provides a **‚ÄúRun Full Analysis‚Äù** button. When clicked, it performs the entire process by cleaning the text, classifying sentiment, and generating summaries for all articles‚Äîoffering a comprehensive overview of how tourism is being discussed.
Butthis automated analysis is time-consuming to run within the Streamlit interface, but it provides a valuable high-level snapshot for large batches of articles.

```python
def run_streamlit():
    # Kill any existing process on port 8501
    os.system('kill -9 $(lsof -t -i:8501) 2>/dev/null')

    # Start Streamlit in the background
    os.system('streamlit run app.py --server.port 8501 &')

    # Wait for the server to start
    time.sleep(5)

    try:
        # Create a ngrok tunnel to the Streamlit app
        tunnel = ngrok.connect(
            addr="8501",  # The port Streamlit is running on
            proto="http",  # Protocol to use
            bind_tls=True  # Enable HTTPS
        )
        print(f'üëâ Your Streamlit app is accessible at: {tunnel.public_url}')
    except Exception as e:
        print(f"Error creating tunnel: {str(e)}")
        # Stop the server in case of error
        os.system('kill -9 $(lsof -t -i:8501) 2>/dev/null')

run_streamlit()
```
